### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
array = np.random.randint(1, high=50, size=10, dtype=int)
media = array.mean()
desvpad = array.std()

print(array)
print(media)
print(desvpad)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
array = np.random.rand(10)
print(array)
array[array<=0.5] = np.sqrt(array[array<=0.5])

print("raiz")
print(array)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3,3,3)

print(matriz)

np.amax(matriz, axis=1)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5,5)

def res(x):
  media = (x[0] + x[1] + x[2] + x[3] + x[4])/5
  return [media]

medias = np.apply_along_axis(res, axis=1, arr=matriz)
print("original")
print(matriz)
print("medias")
print(medias)
print("subtracao")
print(matriz-medias)
```

```
matriz = np.random.rand(5,5)

media = np.mean(matriz, axis=1)

print(matriz - media[:,np.newaxis])
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4,4)
print("matriz original")
print(matriz)
print("\n medias")
print("\t juntas")
print(np.mean([matriz.diagonal(),np.diagonal(np.fliplr(matriz))]))
print('\t maior')
print(np.mean(matriz.diagonal()))
print('\t menor')
print(np.mean(np.diagonal(np.fliplr(matriz))))
print("\n desvios")
print("\t juntas")
print(np.std([matriz.diagonal(),np.diagonal(np.fliplr(matriz))]))
print('\t maior')
print(np.std(matriz.diagonal()))
print('\t menor')
print(np.std(np.diagonal(np.fliplr(matriz))))
```
